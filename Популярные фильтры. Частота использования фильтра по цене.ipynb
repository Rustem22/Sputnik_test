```python
import pandas as pd
data = pd.read_csv('/Users/rustem/Desktop/events filters к тестовому Product Analyst Q1 2024.csv')
```

```python
# Удаление столбца event_label
data.drop('event_label', axis=1, inplace=True)
```

```python
data
```

```python
# Удаление столбца event_category
data.drop('event_category', axis=1, inplace=True)
data
```

```python
# Сгруппировать данные по городам и суммировать метрики
grouped_data = data.groupby('event_action').agg({
    'total_events': 'sum',
    'unique_events': 'sum'
}).reset_index()

grouped_data
```

```python
# Вопрос 3. Топ-10 разделов фильтров.Сортировка данных по столбцу unique_events в порядке убывания
top_event_action_by_unique_events = grouped_data.sort_values(by='unique_events', ascending=False)

top_event_action_by_unique_events.head(10)
```

```python
# Топ-10 разделов фильтров.Сортировка данных по столбцу total_events в порядке убывания
top_event_actions_by_total_events = grouped_data.sort_values(by='total_events', ascending=False)
top_event_actions_by_total_events.head(10)
```

```python
# Вопрос 4. Как часто люди пользуются выбором цен? Выберем только фильтры по ценам и положим значения в список events для фильтрации
events_to_filter = ['Page Visit', 'price_first', 'price_second', 'price_third']

# Фильтрация данных
filtered_events = grouped_data[grouped_data['event_action'].isin(events_to_filter)]

# Вывод результата
print(filtered_events)
```

```python
# Сумма total_events и unique_events для фильтров price_first, price_second, price_third
total_events_filters = filtered_events[filtered_events['event_action'].isin(['price_first', 'price_second', 'price_third'])]['total_events'].sum()
unique_events_filters = filtered_events[filtered_events['event_action'].isin(['price_first', 'price_second', 'price_third'])]['unique_events'].sum()

# Сумма total_events и unique_events для страницы Page Visit
total_events_page_visit = filtered_events[filtered_events['event_action'] == 'Page Visit']['total_events'].sum()
unique_events_page_visit = filtered_events[filtered_events['event_action'] == 'Page Visit']['unique_events'].sum()

# Доля суммы total_events и unique_events по фильтрам от общего количества для страницы Page Visit
total_events_ratio = total_events_filters / total_events_page_visit
unique_events_ratio = unique_events_filters / unique_events_page_visit

# Вывод результатов
print("Сумма общего использования фильтров цен:", total_events_filters)
print("Сумма уникальных пользователей использовавших фильтры цен:", unique_events_filters)
print("Сумма общего посещения страницы Page Visit:", total_events_page_visit)
print("Сумма уникальных пользователей посетивших страницу Page Visit:", unique_events_page_visit)
print("Доля общего количества раз использования фильтра цен от общего количества заходов на страницу:", total_events_ratio)
print("Доля уникальных пользователей, использовавших фильтры цен от общего числа уникальных пользователей зашедших на страницу:", unique_events_ratio)
```
