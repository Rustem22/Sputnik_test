```python
import pandas as pd
```

```python
data = pd.read_csv('/Users/rustem/Desktop/events filters к тестовому Product Analyst Q1 2024.csv')
```

```python
data
```

```python
import re

# Функция для извлечения названия города из event_label
def extract_city(event_label):
    # Используем регулярное выражение для поиска названия города
    match = re.search(r'([^/]+)', event_label)
    if match:
        return match.group(1).strip()  # Возвращаем найденное название города без лишних пробелов
    else:
        return 'Unknown'  # Если название города не найдено, возвращаем 'Unknown'

# Применяем функцию extract_city к столбцу event_label и создаем новый столбец city
data['city'] = data['event_label'].apply(extract_city)

# Выводим первые строки данных для проверки
data.head(10)
```

```python
# Удаление столбца event_label
data.drop('event_label', axis=1, inplace=True)
data.head(10)
```

```python
data = data[data['event_action'] == 'price_button_submit']

data
```

```python
# Сгруппировать данные по городам и суммировать метрики
grouped_data = data.groupby('city').agg({
    'total_events': 'sum',
    'unique_events': 'sum'
}).reset_index()

data.head(10)
```

```python
grouped_data
```

```python
# Сумма значений столбца unique_events
total_unique_events_1 = grouped_data['unique_events'].sum()

# Вывод результата
print("Сумма значений столбца 'unique_events':", total_unique_events_1)
```

```python
# Сортировка данных по столбцу unique_events в порядке убывания
top_cities = grouped_data.sort_values(by='unique_events', ascending=False)

# Выбор первых 50 городов, где фильтры популярны
top_10_cities = top_cities.head(50)

# Вывод результата
top_50_cities
```
